## Data preprocessing

```python
import pandas as pd
```

```python
df = pd.read_csv(r'../data/clean_flipkart_data')
```

```python
df['product_names'] = df['product_name']
df.head()
```

```python
pre_df = df.rename(columns = {'product_name':'products'}, inplace = False)
pre_df.head()
```

```python
import nltk
```

```python
nltk.download('popular')
```

```python
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import PorterStemmer
stop_words = stopwords.words('english')
porter = PorterStemmer()
```

```python
# function to process text
def process_text(input_text):
    tokens = word_tokenize(input_text)
    no_stop = [w for w in tokens if w not in stop_words]
    no_punc = [ch for ch in no_stop if ch.isalpha()]
    stemmed = [(porter.stem(w)) for w in no_punc]
    processed_text = ' '.join(w for w in stemmed)
    return processed_text
```

```python
# applying function
pre_df['products'] = pre_df['products'].apply(process_text)
pre_df['product_category_tree'] = pre_df['product_category_tree'].apply(process_text)
pre_df['description'] = pre_df['description'].astype('str').apply(process_text)
pre_df['brand'] = pre_df['brand'].astype('str').apply(process_text)
```

```python
pre_df.head(10)
```

```python
# combining features
pre_df['combined_features'] = pre_df.products + pre_df.product_category_tree + pre_df.description + pre_df.brand
pre_df.head()
```

```python
pre_df.to_csv('preprocessed_flipkart_data', index = False)
```
